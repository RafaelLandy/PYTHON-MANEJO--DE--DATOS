ESTAS SON LAS LIBRERIAS NECESARIAS

```python
import os
import pandas as pd
import numpy as np
```

Lee los datos de una archivo csv y convierte en dataframe agrega 2 campos con valor de 1 que se usa como contadores

```python
datos_df=pd.read_csv("C:/visual_studio/datos_csv/indice_edificabilidad.csv") 

datos_df['porc']=1 
datos_df['cant_pre']=1
```

Realiza agrupaciones en 2 niveles

```python
grupo2= datos_df.groupby(['CODIGO',]) 
grupo1 = datos_df.groupby(['CODIGO','RANGO_AREA']) 
```

Realiza suma de los cotadores y del campo IE MAX, que en este caso es un valor numerico

```python
suma1= grupo1['porc'].apply(np.sum)
suma2= grupo2['cant_pre'].apply(np.sum)
suma_edf=grupo1['I_E_MAX'].apply(np.sum)
```

Formula para obtener lo deseado, en este caso el promedio de I E MAX, agrupado por nivel 1 y nivel 2

```python
df=suma_edf/suma1 
```

Crea archivos de salida intermedios previo a la salida definitiva

```python
df.to_csv('C:/visual_studio/salidas/porc_aprove.csv',index=True)# el directorio siempre se debe poner con est
suma1.to_csv('C:/visual_studio/salidas/suma1.csv',index=True)

# Lee los archivos de salida
f1=pd.read_csv("C:/visual_studio/salidas/porc_aprove.csv")
f2=pd.read_csv("C:/visual_studio/salidas/suma1.csv")

#Une los archivos de la salida intermedia
merged_df = pd.merge(f1, f2)
#Crea el archivo final de salida
merged_df.to_csv('C:/visual_studio/salidas/Carli333_resultad_rango_edif.csv',index=True)
print('LISTO')
```
